```python
#exercicio 5

def perfil_pessoa(nome, idade):
  letras_nome = len(nome)
  maiusculo = nome.upper()
  maior_idade = idade >= 18
  return nome, maiusculo, letras_nome, maior_idade

nome = 'Gabi'
idade = 23
resultado = perfil_pessoa(nome, idade)
print(f'{resultado[0]}, {resultado[1]}, {resultado[2]}, {resultado[3]}')
```
