```python
#exercicio 1
import math
peso1 = '80.68'
peso1 = round(float(peso1), 1)
peso2 = math.trunc(peso1)
print(f'{peso1}, {peso2} e {peso1:.3f}')
```

```python
#exercicio 2
A = True
B = False

C = A or B
D = A and B
E = not(B)

print(f'{C}, {D} e {E}')
```

```python
#exercicio 3
temperatura = 27

if temperatura < 0:
  print('Muito frio!')
elif (temperatura < 15):
  print('Frio')
elif (temperatura < 27):
  print('Calor')
else:
  print('Muito calor')
```

```python
#exercicio 4
import math
distancia = '90.4734'
distancia1 = round(float(distancia), 2)
distancia2 = math.trunc(distancia1)
distancia3 = math.ceil(distancia1)
distancia4 = math.floor(distancia1)
print(f'{distancia1}, {distancia2}, {distancia3} e {distancia4}')
```
